# 1. NumPy

````python
import numpy as np
````

NumPy is typically imported as the alias "np". This is because we use it so much that we'd like a short identifier.

Now, let's take a look at the most common NumPy operations.

## 1.1. np.arange

````python
np.arange(5)
````

`np.arange` creates a list of integers from 0 to the specified number, exclusive. It is similar to the `range` function in Python.

````python
x = np.arange(5)

# What do you expect this to output?
2 * x
````

When you apply an operation like *, /, +, -, or % between an array and a scalar, the operation is applied element-wise. In the above example, each element is doubled.

## 1.2. np.zeros, np.ones, np.array, ==, np.all

````python
np.zeros((5, 3))
````

`np.zeros` creates a multidimensional array of zeros. The parameter we pass to it is called "shape." For a 2D shape, the first number (5) is the number of rows and the second number (3) is the number of columns.

In math, we call this a matrix, in particular a 5 x 3 matrix.

````python
np.ones((5, 3))
````

`np.ones` does the same thing, except it fills the matrix with ones instead of zeros.

````python
np.array([0, 1, 2, 3, 4])
````

`np.array` converts a Python array into a NumPy array. Note that this is slower than using other NumPy operations such as `arange` and `zeros`. The above array is equivalent to `np.arange(5)`, which brings us to our next topic, comparison.

````python
np.array([0, 1, 2, 3, 4]) == np.arange(5)
````

The == operator compares an array element-wise, yielding a boolean per element.

````python
x = np.array([0, 1, 2, 3, 4]) == np.arange(5)

np.all(x)
````

If we want to check that every element is equal, we can use `np.all`.

````python
x = np.array([0, 1, 2, 3, 4]) == np.arange(5)

1 * x
````

Sometimes it's useful to represent True as 1 and False as 0. To do this, we simply multiply the array by 1. NumPy interprets this as casting a boolean array to an integer array.

## 1.3. Indexing

### 1.3.1. Indexing a single value

We often need to retrieve only part of an array. To do this, we can use indexing.

````python
x = np.arange(10)
x
````

````python
x[0]
````

`x[0]` give us the first element of x

````python
x[4]
````

`x[4]` gives us the 5th element of x

````python
x[-1]
````

`x[-1]` gives us the last element of x

### 1.3.2. Indexing multiple values

We often wish to index multiple values at a time. To do this, we can use slices.

````python
x = np.array([
    [5, -3, 1],
    [1, 5, 9]
])

x[0, :]
````

`x[0, :]` gives us the first row of x. The first number, 0, specifies the row. The second value, :, means every element in that row.

````python
x[:, -1]
````

Similarly, `x[:, -1]` gives us the last column of x.

````python
x[0, 0:2]
````

`x[0, 0:2]` selects the first row and gives us the first 2 elements of that row. We can also write this as `x[0, :2]`, and NumPy will infer the 0 start of the slice.

## 1.4. Broadcasting

Sometimes, we want to set every element of an array to a single value.

````python
x = np.ones(10)
x
````

What if we want to set the first 3 elements to -5?

````python
x[:3] = -5
x
````

`x[:3]` has 3 elements while -5 is a single element. But NumPy infers that you want to set all the values of the left hand side to the value in the right hand side.

This is called broadcasting. If possible, NumPy repeats the shape of the left hand side to fill the left hand side.

````python
x = np.zeros((4, 10))
x
````

````python
x[1:3, 1:9] = np.arange(8)
x
````

Here is another example of broadcasting. The shape of the left hand side is (2, 8). The shape of the right hand side is just (8,).

NumPy will implicitly prepend 1's to the shape such that the length of the shape tuple is equal. It will then copy the values across the newly created channels until the shapes themselves are equal.

## 1.5. Reshaping, indexing via lists

````python
x = np.arange(9)
x.reshape((3, 3))
````

We oftentimes want to change the shape of arrays. The above is an example.

````python
x = np.arange(100).reshape(10, 10)
x
````

In addition to selecting multiple elements via slices, we can also select via lists.

````python
row_indexes = [0, 4, 5, -1]
x[row_indexes, :]
````

The above selects the first, 5th, 6th, and last rows of x.

# 2. MatPlotLib

Now you have the basics of NumPy under your belt. Let's turn our attention to plotting.

We often need to make graphs in machine learning, whether to show loss curves, accuracy benchmarks, etc. So let's learn and practice.

The most popular library for plotting in Python is MatPlotLib.

````python
from matplotlib import pyplot as plt
````

This is how matplotlib is typically imported. Sometimes you'll see `import matplotlib.pyplot as plt`, which is equivalent.

````python
x = np.arange(10)
y = np.arange(10)

plt.plot(x, y)
plt.show()
````

Here we have our first plot. What's happening?

Essentially, we give MatPlotLib a list of (x, y) pairs. It plots each (x, y) point and connects them with line segments.

Here, it's plotting the points (0, 0), (1, 1), (2, 2), ..., (9, 9).

````python
x = np.arange(10)
y = np.sin(x)

plt.plot(x, y)
plt.show()
````

Here you can see what's going on. `np.sin` will take the sine of every element of x. Since we're plotting (0, sin(0)), (1, sin(1)), etc. and connecting them with line segments, the plot is very choppy.

We can fix this by sampling more points.

````python
x = np.linspace(0, 1, 101)
x
````

`np.linspace` is a commonly used function to create an evenly spaced interval. You specify the start value, end value, and number of elements.

In this case, we specify 101 evenly spaced values between 0 and 1.

````python
x = np.linspace(-5, 5, 101)
y = np.sin(x)

plt.plot(x, y)
plt.show()
````

Our sine graph looks much better now! Let's add some labels.

````python
x = np.linspace(-5, 5, 101)
y = np.sin(x)

plt.plot(x, y)

plt.xlabel("x")
plt.ylabel("sin(x)")
plt.title("Sine of x")
plt.show()
````

Sometimes, we want to plot multiple curves on the same graph. To do this, we can specify a label.

````python
x = np.linspace(-5, 5, 101)
y = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y)
plt.plot(x, y2)

plt.xlabel("x")
plt.ylabel("y")
plt.title("Sine and cosine of x")
plt.show()
````

We simply call `plt.plot` twice. But how do we distinguish which is which?

````python
x = np.linspace(-5, 5, 101)
y = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y, label="sin(x)")
plt.plot(x, y2, label="cos(x)")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Sine and cosine of x")
plt.legend()
plt.show()
````

We simply pass `label="..."` to `plt.plot`, then call `plt.legend`.

````python
fig, ax = plt.subplots(1, 2)
````

You'll often see the above used. This is used to create multiple plots.

fig stands for "figure". it's essentially the window. you can use it to change the viewport size, save the figure, and so on.

ax is an array of all the axes created. you can use `ax[0].plot` instead of `plt.plot` to plot on the first graph.

````python
fig, ax = plt.subplots(1, 2)

ax[0].plot(x, y)
ax[1].plot(x, y2)

plt.show()
````

We can add labels, just like before.

````python
fig, ax = plt.subplots(1, 2)
fig.set_figwidth(15)

ax[0].plot(x, y)
ax[0].set_xlabel("x")
ax[0].set_ylabel("sin(x)")
ax[0].set_title("sin(x)")

ax[1].plot(x, y2)
ax[1].set_xlabel("x")
ax[1].set_ylabel("cos(x)")
ax[1].set_title("cos(x)")

plt.show()
````
